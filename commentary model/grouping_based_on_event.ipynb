```python
import pandas as pd
```

```python
path='Data/cleansed_data/clean_commentary.csv'
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
groups = df.groupby('score')

for group_name, group_df in groups:
    print('Group Name:', group_name)
    print(group_df.head())
    #save each DataFrame to a file
    filename = f'output/eventwise_csv/{group_name}.csv'
    group_df.to_csv(filename, index=False)
```
