```python
import pandas as pd
import numpy as np
```

```python
file_url = 'Data/commentary.csv'
df = pd.read_csv(file_url)
```

```python
df_cleaned = df.copy()
dup = df_cleaned.duplicated()
df_cleaned[dup]
```

```python
df_cleaned['Commentary_new'] = np.nan
df_cleaned['Team_new'] = df_cleaned['Team'].str.split(' ').str[0]
df_cleaned[['P2', 'P1']] = df_cleaned.batsman.str.split(" to ", expand = True)
df_cleaned[['P1_FN', 'P1_LN']] = df_cleaned['P1'].str.split(" ", 1, expand = True)
df_cleaned[['P2_FN', 'P2_LN']] = df_cleaned['P2'].str.split(" ", 1, expand = True)
```

```python
df_cleaned.sample(5)
```

```python
print(df_cleaned['Team_new'].nunique())
print(df_cleaned['Team_new'].unique())
print(df_cleaned['P1'].nunique())
print(df_cleaned['P2'].nunique())
print(df_cleaned['score'].nunique())
print(df_cleaned['score'].unique())
```

```python
for ind in df_cleaned.index:
  df_cleaned['Commentary_new'][ind] = df_cleaned['Commentary'][ind].replace(df_cleaned['P1'][ind], '{{BATSMAN}}')
  df_cleaned['Commentary_new'][ind] = df_cleaned['Commentary_new'][ind].replace(df_cleaned['P1_FN'][ind], '{{BATSMAN}}')
  if (df_cleaned['P1_LN'][ind] != None) :
    df_cleaned['Commentary_new'][ind] = df_cleaned['Commentary_new'][ind].replace(df_cleaned['P1_LN'][ind], '{{BATSMAN}}')
  df_cleaned['Commentary_new'][ind] = df_cleaned['Commentary_new'][ind].replace(df_cleaned['P2'][ind], '{{BOWLER}}')
  df_cleaned['Commentary_new'][ind] = df_cleaned['Commentary_new'][ind].replace(df_cleaned['P2_FN'][ind], '{{BOWLER}}')
  if (df_cleaned['P2_LN'][ind] != None) :
    df_cleaned['Commentary_new'][ind] = df_cleaned['Commentary_new'][ind].replace(df_cleaned['P2_LN'][ind], '{{BOWLER}}')
  df_cleaned['Commentary_new'][ind] = df_cleaned['Commentary_new'][ind].replace(df_cleaned['Team_new'][ind], '{{TEAM}}')
```

```python
df_cleaned.sample(5)
```

```python
df_cleaned['P1_FN_count'] = df_cleaned['P1_FN'].str.len()
df_cleaned['P1_LN_count'] = df_cleaned['P1_LN'].str.len()
df_cleaned['P2_FN_count'] = df_cleaned['P2_FN'].str.len()
df_cleaned['P2_LN_count'] = df_cleaned['P2_LN'].str.len()
```

```python
df_cleaned.sample(5)
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.shape
```

```python
count_1 = 0
count_2 = 0
count_3 = 0
count_4 = 0

for ind in df_cleaned.index:
  if df_cleaned['P1_FN_count'][ind] == 1 :
    count_1 = count_1 + 1
  if df_cleaned['P1_LN_count'][ind] == 1 :
    count_2 = count_2 + 1
  if df_cleaned['P2_FN_count'][ind] == 1 :
    count_3 = count_3 + 1
  if df_cleaned['P2_LN_count'][ind] == 1 :
    count_4 = count_4 + 1
print(count_1, count_2, count_3, count_4)
```

```python
df_cleaned.drop(df_cleaned[(df_cleaned['P1_FN_count'])==1].index, inplace=True)
df_cleaned.drop(df_cleaned[(df_cleaned['P2_FN_count'])==1].index, inplace=True)
```
